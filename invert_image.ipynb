```python
import cv2 as cv
import matplotlib.pyplot as plt
import json
import numpy as np
```

### read image

```python
photo = 'chrisma.PNG'
img = cv.imread(photo)
assert img is not None, "No image availbe" 
```

### Get collision points


```python
# Convert image to grayscale
img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
plt.imshow(img, 'gray')
plt.axis('off')
plt.show()

# Convert image to blurr
blurr = cv.GaussianBlur(img, (7,7), cv.BORDER_DEFAULT)
plt.imshow(blurr, cmap='gray')
plt.axis('off')
plt.title('Blurred')
plt.show()
```

### Creatin edge nodes

```python
canny = cv.Canny(img, 125, 175)
plt.imshow(canny, cmap='gray')
plt.axis('off')
plt.show()
```

### Convert to json file

```python
# Get x,y points
points = np.column_stack(np.where(canny > 0))
points = [(int(x), int(y)) for y, x in points]
# import to json
with open("points.json", "w") as f:

    json.dump(points,f)

```
